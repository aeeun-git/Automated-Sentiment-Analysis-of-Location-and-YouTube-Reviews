```python
import pandas as pd

# 1. 빈 리스트 생성
all_data = []

# ==========================================
# (1) NSMC 처리
# ==========================================
print("Processing NSMC...")
nsmc = pd.read_table('ratings_train.txt') # 파일 경로 수정 필요
nsmc = nsmc[['document', 'label']] # 필요한 컬럼만 선택
nsmc.columns = ['text', 'label']   # 이름 통일
nsmc.dropna(inplace=True)          # 빈 값 제거
all_data.append(nsmc)

# ==========================================
# (2) Kaggle POI 처리 (예시)
# ==========================================
print("Processing Kaggle POI...")
# CSV 파일이라고 가정
poi = pd.read_csv('kakao_reviews.csv') 
# 컬럼명이 'review_txt', 'score' 라고 가정 (실제 파일 열어서 확인 후 수정하세요!)
# 별점 4점 이상 긍정, 2점 이하 부정 처리 예시
poi_pos = poi[poi['score'] >= 4].copy()
poi_pos['label'] = 1
poi_neg = poi[poi['score'] <= 2].copy()
poi_neg['label'] = 0

poi_final = pd.concat([poi_pos, poi_neg])
poi_final = poi_final[['review_txt', 'label']] # 컬럼 선택
poi_final.columns = ['text', 'label']          # 이름 통일
all_data.append(poi_final)
```
